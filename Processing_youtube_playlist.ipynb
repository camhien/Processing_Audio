```python
def split_audio(split_path, srt_path, mono_wav_path, _id):
    subs = pysrt.open(srt_path)
    # Start and End time 
    start_time = []
    end_time = []
    id_of_audio = _id
    for sub in subs:
        start = sub.start.to_time()
        start_millisecond = int(start.hour*3600000 + start.minute*60000 + start.second*1000 + start.microsecond*0.001)
        end = sub.end.to_time()
        end_millisecond = int(end.hour*3600000 + end.minute*60000 + end.second*1000 + end.microsecond*0.001)
        duration = (end_millisecond - start_millisecond)/1000
        name = f'{id_of_audio}_{start_millisecond}_{end_millisecond}'
        name = os.path.join(split_path, name)
        !ffmpeg -ss {start} -i {mono_wav_path} -t {duration} {name}.wav
    
def processing_one_id(_id, output_dir):
        path = os.path.join(output_dir, _id)
        print(path)
        os.mkdir(path)
        wav_path = os.path.join(path, _id + ".wav")
        print(wav_path)
        !youtube-dl -x --audio-format wav -i {_id} -o {wav_path}
        mono_wav_path = os.path.join(path, _id + ".mono.wav")
        print(mono_wav_path)
        !ffmpeg -i {wav_path} -ac 1 -ar 16000 {mono_wav_path}
        srt_path = os.path.join(path, _id + ".srt")
        print(srt_path)
        !autosub {mono_wav_path} -S vi -D vi -o {srt_path}
        split_path = os.path.join(path, "splits")
        os.mkdir(split_path)
        split_audio(split_path, srt_path, mono_wav_path, _id)
        
        

def processingAudio(id_of_playlist, output_dir):
    
    list_of_ids = !youtube-dl --get-id PLIuJsuIqMEjdtfslVTUT8iwGv7_e1J7yX
    
    for _id in list_of_ids:
        processing_one_id(_id, output_dir)
        
processingAudio('PL4wNiKPNg9RjlVwUK32q7JzNbKcgJRDcj', 'D:\\')
        
    
```
